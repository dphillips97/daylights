```python
# load data, text file with tab delimiter
df = pd.read_csv('edited_data.txt', delimiter = '\t')
df
```

```python
# create function
def decimal_time(t):
    try:
        t = t.split(':')
        hrs = float(t[0])
        min = float(t[1]) / 60
        return str(round(hrs + min, 3))
    except:
        return '0'
```

```python
cols = list(df.columns)
```

```python
# test function, return as string
decimal_time('12:28')
```

```python
cols
```

```python
for col in cols:
    df[col] = df[col].apply(decimal_time)
```
